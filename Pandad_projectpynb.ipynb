```python
df = pd.read_csv("/content/noisy_covid_data.csv")
print(df)
```

```python
dr = df[df["Confirmed"]>500000]
print(dr)
```

# ***USE CASE 2***

# **GROUP DATA BY COUNTRY**

```python
dr = df.groupby("Country").sum()
print(dr)
```

# **RANK CONFIRMED BY CASES**

```python
dr.sort_values("Confirmed", ascending = False)
```

# **FATILITY RATE**

```python
f = (dr["Deaths"]/dr["Confirmed"])*100
print(f)
```

```python
f.fillna(0)
```

# ***USE CASE 3***

# **FILTER DATA BY COUNTRY**

```python
r = df[df["Country"] == "India"]
print(r)
```

```python
r.fillna(0)
```

# **GROUP BY DATE**

```python
df.groupby(["Date","Confirmed","Recovered"])["Deaths"].sum()
```

# **VISUALIZE TRENDS**

```python
import matplotlib.pyplot as plt
df.plot(kind = "line")
plt.xlabel("Deaths")
plt.ylabel("Confirmed")
plt.show()
```

# ***USE CASE 4***

# **RESAMPLE DATA**

```python
df.resample("W").sum()
```

# **CALCULATING THE ROLLING AVERAGE**

```python
r = df.groupby(["Country","Date"])["Confirmed"].rolling(1).sum()
print(r)
```

```python
r.fillna(0)
```

# **VISUALIZING THE ROLLING AVERAGE**

```python
r.plot(kind = "line")
plt.xlabel("Date")
plt.ylabel("Confirmed")
plt.show()
```
